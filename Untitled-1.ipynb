```python
def transform_dataset(df):
    # Load the dataset

    # Filter out rows where 'related' is NaN
    filtered_dataset = df.dropna(subset=['related'])

    # Create a new DataFrame for the transformed data
    transformed_data = []
    for index, row in filtered_dataset.iterrows():
        name = row['name']
        related_names = row['related'].split(',')
        for related_name in related_names:
            transformed_data.append([name, related_name])

    # Converting the list to a DataFrame
    transformed_df = pd.DataFrame(
        transformed_data, columns=['name', 'relation to'])
    transformed_df.to_csv('./transformed_dataset.csv', index=False)
    return transformed_df


# Example usage
if __name__ == "__main__":
    file_path = df
    transformed_df = transform_dataset(file_path)
    # Display the first few rows of the transformed DataFrame
    print(transformed_df.head())
```
